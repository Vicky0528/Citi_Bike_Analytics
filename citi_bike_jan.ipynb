```python
import pandas as pd
```

```python
jan2018 = "data/2018/201801-citibike-tripdata.csv"
jan2019 = "data/2019/201901-citibike-tripdata.csv"
jan2020 = "data/2020/202001-citibike-tripdata.csv"
```

```python
jan18_df = pd.read_csv(jan2018)
jan18_df.head()
```

```python
jan19_df = pd.read_csv(jan2019)
jan19_df.head()
```

```python
jan20_df = pd.read_csv(jan2020)
jan20_df.head()
```

```python
jan_df = jan18_df.append([jan19_df,jan20_df],ignore_index = True)
jan_df.head()
```

```python
# Save Combined DataFrames to CSV
jan_df.to_csv("data/jan.csv", index=False, header=True)
```
